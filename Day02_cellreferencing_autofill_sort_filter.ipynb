# ‚≠ê 1. Cell Referencing in Excel

Cell referencing tells Excel how to treat the cells when formulas are copied.

## ‚úî 1.1 Relative Reference (A1)
- Changes when copied.
- Default type.
Example:  
If A2 = 10, A3 = 20  
Formula in B2: `=A2 + 5`  
Copy down ‚Üí B3 becomes `=A3 + 5`

---

## ‚úî 1.2 Absolute Reference ($A$1)
- Does NOT change when copied.
- Use `$` before column and row.
Example:  
`=$A$2 * B2`

---

## ‚úî 1.3 Mixed Reference ($A1 or A$1)
- Either column is fixed or row is fixed.
Examples:  
`=$A1` ‚Üí column fixed  
`=A$1` ‚Üí row fixed

# ‚≠ê 2. Autofill

Autofill automatically fills:
- Numbers (1, 2, 3‚Ä¶)
- Dates (1-Jan, 2-Jan‚Ä¶)
- Days (Mon, Tue‚Ä¶)
- Months (Jan, Feb‚Ä¶)
- Formulas  
- Custom patterns

Use **bottom-right small box (fill handle)**.

---

# ‚≠ê 3. Sorting

Sorting rearranges data.

Types:
- **Ascending** (A‚ÜíZ or 0‚Üí9)
- **Descending** (Z‚ÜíA or 9‚Üí0)
- **Custom Sort**  
- **Multi-Level Sort** (Sort by Name, then by Age)

---

# ‚≠ê 4. Filtering

Filtering hides data that doesn't match conditions.

Common filters:
- Text Filter (Begins With, Contains)
- Number Filter (Greater Than, Between)
- Date Filter (This Week, Last Month)
- Top 10 Filter
- Clear Filter

# ‚≠ê 7. Interview Questions

---

### **Q1. What is the difference between relative, absolute, and mixed referencing?**  
Relative ‚Üí changes  
Absolute ‚Üí fixed  
Mixed ‚Üí partly fixed

---

### **Q2. How do you lock a cell reference?**  
Use **$** symbol ‚Üí `$A$1`.

---

### **Q3. Why do we use absolute referencing?**  
To keep a constant value (like tax %, exchange rate) in formulas.

---

### **Q4. What is Autofill used for?**  
To quickly fill sequences, formulas, or patterns.

---

### **Q5. What are the types of sorting in Excel?**  
Ascending, Descending, Custom, Multi-level.

---

### **Q6. What is filtering in Excel?**  
Hides rows that don‚Äôt match conditions.

---

### **Q7. Difference between Sort and Filter?**  
Sort rearranges data; Filter hides irrelevant data.

---

### **Q8. What is a custom sort?**  
Sorting based on a user-defined list (e.g., Jan ‚Üí Feb ‚Üí Mar).

---

### **Q9. What is the purpose of multi-level sorting?**  
Sort by more than one column (e.g., Category ‚Üí Price).

---

### **Q10. Can Autofill copy formulas?**  
Yes ‚Äî formulas adjust automatically as per referencing type.
