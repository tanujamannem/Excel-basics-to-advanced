**Q1. What is Excel and how is it used in data analysis?**

Excel is a spreadsheet tool used to store, clean, analyze, and visualize data. Data analysts use Excel for initial exploration, validation, quick calculations, and reporting before moving to tools like Power BI or SQL.

---

**Q2. What are workbooks and worksheets?**

A workbook is an Excel file, while worksheets are individual sheets inside the workbook where data is stored and analyzed.

---

**Q3. What are rows and columns in Excel?**

Rows run horizontally and represent records, while columns run vertically and represent attributes or fields.

---

**Q4. Explain cell referencing in Excel.**

Cell references identify cells in formulas. They can be relative, absolute, or mixed, depending on whether they change when copied.

---

**Q5. What is relative cell referencing?**

Relative references change automatically when a formula is copied to another cell.

---

**Q6. What is absolute cell referencing?**

Absolute references remain fixed using the dollar sign and are used for constants like tax rates or thresholds.

---

**Q7. What is mixed cell referencing?**

Mixed references lock either the row or the column while allowing the other to change.

---

**Q8. What are Excel formulas and functions?**

Formulas perform calculations, while functions are predefined formulas that simplify common calculations.

---

**Q9. What are the most commonly used Excel functions for analysts?**

SUM, AVERAGE, COUNT, COUNTIF, SUMIF, IF, AND, OR, VLOOKUP, XLOOKUP, INDEX, MATCH, TEXT, and DATE functions.

---

**Q10. Difference between COUNT, COUNTA, and COUNTBLANK?**

COUNT counts numeric values, COUNTA counts non-empty cells, and COUNTBLANK counts empty cells.

---

**Q11. Explain IF function with business usage.**

IF applies conditional logic to classify or validate data, such as marking sales as High or Low.

---

**Q12. What is the use of AND and OR functions?**

AND checks if all conditions are true, OR checks if any one condition is true.

---

**Q13. What are text functions in Excel?**

Text functions manipulate text data, such as LEFT, RIGHT, MID, LEN, TRIM, and CONCAT.

---

**Q14. Why is TRIM important in data cleaning?**

TRIM removes extra spaces that cause mismatches in lookups and filtering.

---

**Q15. What are date and time functions in Excel?**

Functions like TODAY, NOW, YEAR, MONTH, and DAY help analyze time-based data.

---

**Q16. What are lookup functions in Excel?**

Lookup functions fetch values from another table based on a matching key.

---

**Q17. Explain VLOOKUP and its limitation.**

VLOOKUP searches vertically but requires the lookup column to be on the left and breaks if columns move.

---

**Q18. Why is XLOOKUP preferred over VLOOKUP?**

XLOOKUP supports flexible lookups, handles missing values, and works in any direction.

---

**Q19. Difference between INDEX-MATCH and VLOOKUP?**

INDEX-MATCH is more flexible, faster on large data, and does not depend on column position.

---

**Q20. What is approximate vs exact match in lookups?**

Exact match returns precise values, while approximate match returns closest values, often used in grading systems.

---

**Q21. What are Pivot Tables?**

Pivot Tables summarize large datasets quickly using drag-and-drop aggregation.

---

**Q22. Why are Pivot Tables important for analysts?**

They allow fast analysis without writing formulas and support dynamic filtering.

---

**Q23. What are calculated fields in Pivot Tables?**

Calculated fields create new metrics using existing fields inside Pivot Tables.

---

**Q24. Difference between Pivot Table and Power Pivot?**

Pivot Tables are simple, while Power Pivot supports data models and DAX calculations.

---

**Q25. What are slicers in Excel?**

Slicers are visual filters that allow users to interactively filter Pivot Tables.

---

**Q26. Difference between slicers and filters?**

Filters are dropdown-based, while slicers are visual and user-friendly.

---

**Q27. What is Conditional Formatting?**

Conditional Formatting highlights data based on rules to show trends and exceptions.

---

---

**Q29. How do you handle missing values in Excel?**

By filtering, replacing with defaults, or using IF and ISBLANK logic.

---

**Q30. What is Text to Columns used for?**

It splits combined data into separate columns using delimiters.

---

**Q31. What is Power Query?**

Power Query is a tool for data extraction, transformation, and loading.

---

**Q32. Why use Power Query instead of formulas?**

Power Query automates cleaning steps and improves performance.

---

**Q33. What are common Power Query transformations?**

Removing duplicates, changing data types, merging, appending, and splitting columns.

---

**Q34. What does Merge Queries do?**

Merge combines tables based on a common key, similar to SQL joins.

---

**Q35. What does Append Queries do?**

Append stacks tables with the same structure.

---

**Q36. What is Excel Table and its advantages?**

Tables auto-expand, maintain formulas, and improve readability.

---

**Q37. What are charts in Excel?**

Charts visualize data, including column, bar, line, pie, and scatter charts.

---

**Q38. Which chart is best for trend analysis?**

Line charts are best for trends over time.

---

**Q39. What is a dashboard in Excel?**

A dashboard is a visual summary of key metrics using charts, KPIs, and slicers.

---

**Q40. What makes a good Excel dashboard?**

Clear KPIs, minimal clutter, consistent formatting, and interactivity.

---

**Q41. What is data validation?**

Data validation restricts incorrect data entry to maintain quality.

---

**Q42. What is Excel automation?**

Automation reduces repetitive tasks using macros or Power Query.

---

**Q43. What are macros in Excel?**

Macros record and automate repetitive actions using VBA.

---

**Q44. Difference between Macro and Power Query?**

Macros automate actions, Power Query automates data transformation.

---

**Q45. What is What-If Analysis?**

What-If tools analyze different scenarios using Goal Seek and Data Tables.

---

**Q46. What is Goal Seek used for?**

It finds the input value required to achieve a desired result.

**Q47. What are Excel shortcuts and why are they important?**

Shortcuts improve speed and productivity during analysis.

**Q48. How do you optimize Excel performance?**

Use tables, avoid volatile formulas, limit formatting, and use Power Query.

**Q49. What are volatile functions?**

Functions like NOW and TODAY recalculate frequently and slow performance.

---

**Q50. How is Excel used with Power BI?**

Excel is used for data prep and validation before building dashboards in Power BI.

**Q54. How do you check data quality in Excel?**

Using filters, counts, duplicates check, and validation rules.

**Q55. What is structured referencing?**

Using table column names instead of cell ranges in formulas.

---

**Q56. Difference between Excel and Google Sheets?**

Excel is more powerful for advanced analysis and large datasets.

**Q57. Can Excel handle large datasets?**

Yes, but performance depends on optimization and structure.

---

**Q58. When should Excel not be used?**

For very large or real-time datasets, databases or BI tools are better.

---

**Q59. How do you keep Excel files error-free?**

By validating data, documenting logic, and testing calculations.

---

**Q60. Final Excel takeaway**
