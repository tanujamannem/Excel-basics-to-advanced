# â­ 1. Introduction to Conditional Functions

Conditional functions perform calculations **based on criteria**.

Today you will learn:

- IF  
- SUMIF  
- COUNTIF  
- AVERAGEIF  

These are among the MOST common functions used by Data Analysts.

---

# â­ 2. IF Function

### âœ” Syntax  
`=IF(condition, value_if_true, value_if_false)`

### âœ” Examples  
`=IF(A2 > 50, "Pass", "Fail")`  
`=IF(B2="India", 1, 0)`

---

# â­ 3. SUMIF Function

### âœ” Syntax  
`=SUMIF(range, criteria, sum_range)`

### âœ” Example  
`=SUMIF(A2:A10, "Stationery", C2:C10)`

---

# â­ 4. COUNTIF Function

# â­ 5. AVERAGEIF Function

# â­ 8. Interview Questions

---

### **Q1. What is the IF function used for?**  
Checks a condition and returns different values for TRUE/FALSE.

---

### **Q2. What is the difference between SUM and SUMIF?**  
SUM â†’ adds all values  
SUMIF â†’ adds values that meet a condition.

---

### **Q3. Can SUMIF handle multiple conditions?**  
No â†’ For multiple conditions use **SUMIFS** (future topic).

---

### **Q4. What is COUNTIF used for?**  
Counts cells that satisfy a condition (e.g., >50, "Pen").

---

### **Q5. What is the wildcard character in COUNTIF?**  
`*` â†’ matches any number of characters  
`?` â†’ matches a single character

---

### **Q6. Difference between COUNTIF and COUNTIFS?**  
COUNTIF â†’ 1 condition  
COUNTIFS â†’ multiple conditions

---

### **Q7. What is AVERAGEIF used for?**  
Calculates average of values that match a condition.

---

### **Q8. Can IF function be nested?**  
Yes â†’ multiple IFs can be used.  
Example: `=IF(A2>80,"High",IF(A2>50,"Medium","Low"))`

---

### **Q9. What happens if SUMIF criteria doesn't match any value?**  
Result = **0**

---

### **Q10. Can SUMIF work with text criteria?**  
Yes â†’ example: `"Stationery"`, `"<>Kitchen"`, `"P*"`.
