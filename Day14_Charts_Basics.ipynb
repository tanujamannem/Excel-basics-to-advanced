# ‚≠ê 1. Introduction to Charts

Charts are visual representations of data that help users:
- Identify trends
- Compare values
- Communicate insights clearly
- Build dashboards and reports

Excel charts are widely used in reporting and decision-making.

---

# ‚≠ê 2. Types of Charts Covered

## Column Chart
Used to compare values across categories.

## Bar Chart
Similar to column chart but displayed horizontally.

## Line Chart
Used to show trends over time.

## Pie Chart
Used to show proportion or percentage contribution.

## Doughnut Chart
Similar to pie chart with a hollow center.

## Combination Chart
Uses two chart types together (e.g., Column + Line).
