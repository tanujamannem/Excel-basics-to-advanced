# ‚≠ê 1. Introduction to Power Query Transformations

Power Query Transformations are used to **reshape, restructure, and prepare data** for analysis.

Transformations allow you to:
- Change data structure
- Combine datasets
- Normalize data
- Prepare data for Pivot Tables, Power BI, and dashboards

Unlike Excel formulas, transformations are:
- Step-based
- Repeatable
- Automated on refresh

---

# ‚≠ê 2. Split Column (Very Important)

Used when multiple values are stored in one column.

### Common Use Cases:
- Full Name ‚Üí First Name + Last Name
- DateTime ‚Üí Date + Time
- City-State ‚Üí City, State

### How to Split:
Transform ‚Üí Split Column ‚Üí By Delimiter / By Number of Characters

Power Query automatically creates new columns.

---

# ‚≠ê 3. Merge Queries (Join Tables)

Merge is used to **combine columns from another table**, similar to SQL JOIN.

### Join Types:
- Left Outer (MOST USED)
- Right Outer
- Inner
- Full Outer
- Anti Joins

### Common Use Case:
Sales table + Product table

Steps:
Home ‚Üí Merge Queries ‚Üí Select matching columns ‚Üí Choose join type

---

# ‚≠ê 4. Append Queries (Stack Tables)

Append is used to **combine rows** from multiple tables.

### Use When:
- Monthly files need to be combined
- Same column structure exists

Append does NOT match keys, it stacks data vertically.

---

# ‚≠ê 5. Unpivot Columns (CRITICAL FOR ANALYSTS)

Unpivot converts **wide data to long format**.

### Why Unpivot is Important:
- Required for Pivot Tables
- Required for Power BI
- Makes data analysis-ready

### Example:
Sales_Jan | Sales_Feb | Sales_Mar  
‚Üí Month | Sales

Steps:
Select columns ‚Üí Transform ‚Üí Unpivot Columns

---

# ‚≠ê 6. Pivot Columns (Reverse of Unpivot)

Pivot converts long data back to wide format.

Used when:
- You need summary tables
- Reporting format requires wide layout

---

# ‚≠ê 7. Replace Values

Used to standardize data:
- Yes/No ‚Üí Y/N
- NA ‚Üí 0
- Incorrect labels ‚Üí Correct values

Very common cleaning step.

---

# ‚≠ê 8. Change Data Types (Revisited)

Correct data types are CRITICAL:
- Number
- Date
- Text
- Boolean

Wrong data types cause:
- Calculation errors
- Refresh failures
