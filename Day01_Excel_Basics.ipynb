# ‚≠ê 1. Introduction to Excel
Excel is a spreadsheet tool used for data entry, calculations, analysis, dashboards, and reporting.

### Key Components
- **Workbook** ‚Üí Entire Excel file (.xlsx)
- **Worksheet** ‚Üí Individual tabs in a workbook
- **Cell** ‚Üí Smallest unit (A1, B2)
- **Range** ‚Üí Group of cells (A1:C5)
- **Ribbon** ‚Üí Toolbar with menus (Home, Insert, Data)
- **Formula Bar** ‚Üí Displays or edits formulas inside a cell

### ‚úî Data Types in Excel

### ‚úî Operators in Excel

# ‚≠ê 3. Must-Know Basic Formulas
| Function | Use | Example |
|---------|-----|----------|
| SUM | Adds values | `=SUM(C2:C6)` |
| AVERAGE | Mean value | `=AVERAGE(C2:C6)` |
| MIN | Smallest | `=MIN(C2:C6)` |
| MAX | Largest | `=MAX(C2:C6)` |
| COUNT | Counts numbers | `=COUNT(C2:C6)` |
| COUNTA | Counts non-blank cells | `=COUNTA(A2:A6)` |
| CONCAT / & | Joins text | `=A2 & " scored " & C2` |

# ‚≠ê 6.Interview Questions

---

### **Q1. Difference between Workbook and Worksheet?**  
Workbook = Excel file  
Worksheet = Individual sheet inside the file

---

### **Q2. What is a cell reference?**  
The address of a cell, e.g., A1, B2.

---

### **Q3. Types of data in Excel?**  
Text, Numbers, Dates, Boolean, Errors.

---

### **Q4. Difference between COUNT and COUNTA?**  
COUNT ‚Üí only numbers  
COUNTA ‚Üí non-empty cells

---

### **Q5. What is a range?**  
A group of cells, e.g., A1:C5.

---

### **Q6. What is Autofill?**  
Feature that fills patterns, sequences, or formulas automatically.

---

### **Q7. What is the difference between formula and function?**  
Formula = custom calculation  
Function = predefined formula (SUM, AVERAGE)

---

### **Q8. What are common error types?**  
 #DIV/0!, #VALUE!, #REF!, #N/A

---

### **Q9. What is the use of the formula bar?**  
To view and edit formulas in a cell.

---

### **Q10. What is the default Excel file extension?**  
`.xlsx`
