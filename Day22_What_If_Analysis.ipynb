# ‚≠ê 1. What is What-If Analysis?

What-If Analysis is used to:
- Test different possibilities
- See how changes in inputs affect results
- Support business decision-making

Very common use cases:
- Profit planning
- Target sales calculation
- Budget forecasting
- Sensitivity analysis

Excel provides THREE main What-If tools:
- Goal Seek
- Scenario Manager
- Data Tables

---

# ‚≠ê 2. Goal Seek (Most Common & Important)

Goal Seek finds the **input value** required to reach a **desired result**.

Think of it as:
> ‚ÄúI know the answer I want ‚Äî Excel, find the input.‚Äù

---

### When to use Goal Seek:
- Only ONE input variable
- One formula
- One target result

---

### Goal Seek Inputs:
- Set cell ‚Üí Result cell (formula)
- To value ‚Üí Desired output
- By changing cell ‚Üí Input value

---

### Example:
Revenue = Price √ó Quantity

If:
- Price = 50
- Quantity = ?
- Target Revenue = 10,000

Goal Seek finds Quantity automatically.

---

# ‚≠ê 3. Steps to Use Goal Seek

1. Create formula (e.g., Revenue)
2. Go to **Data ‚Üí What-If Analysis ‚Üí Goal Seek**
3. Set Cell ‚Üí Revenue cell
4. To Value ‚Üí Target number
5. By Changing Cell ‚Üí Input cell
6. Click OK

Excel iterates until target is met.

---

# ‚≠ê 4. Limitations of Goal Seek (Interview Question)

- Only one variable
- No constraints
- Cannot save multiple cases

For multiple scenarios ‚Üí use Scenario Manager.

---

# ‚≠ê 5. Scenario Manager (Multiple Scenarios)

Scenario Manager allows you to:
- Save multiple sets of inputs
- Compare different situations
- Switch scenarios instantly

Used when:
- Multiple variables change
- Management wants comparisons

---

### Example Scenarios:
- Best Case
- Worst Case
- Expected Case

Each scenario stores:
- Different input values
- Same formula structure

---

# ‚≠ê 6. Steps to Use Scenario Manager

1. Data ‚Üí What-If Analysis ‚Üí Scenario Manager
2. Click Add
3. Name scenario
4. Select changing cells
5. Enter values
6. Repeat for multiple scenarios
7. Show or generate summary

---

# ‚≠ê 7. Scenario Summary Report

Scenario Summary:
- Auto-generated report
- Compares outputs across scenarios
- Very useful for presentations

Two types:
- Scenario Summary
- Scenario Pivot Table

---

# ‚≠ê 8. Data Tables (Sensitivity Analysis)

Data Tables show:
- How output changes when inputs vary
- Many values at once

Used heavily in:
- Finance
- Forecasting
- Pricing analysis

---

# ‚≠ê 9. One-Variable Data Table

Used when:
- One input variable
- Many input values

Example:
- Different prices
- Resulting revenue

Excel recalculates formula for each value.

---

# ‚≠ê 10. Two-Variable Data Table

Used when:
- Two input variables
- One result formula

Example:
- Price
- Quantity
- Result ‚Üí Revenue

Creates a matrix of outputs.
