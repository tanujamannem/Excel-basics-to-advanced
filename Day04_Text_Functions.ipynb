# ‚≠ê 1. Introduction to Text Functions

Text functions are used to **clean, format, extract, and combine text data**.

They are heavily used in:
- Data cleaning
- Preparing raw datasets
- Creating readable reports

Commonly used by Data Analysts on names, emails, cities, IDs, etc.

---

# ‚≠ê 2. LEFT Function

### ‚úî Syntax  
`=LEFT(text, number_of_characters)`

### ‚úî Example  
`=LEFT(A2,4)`

---

# ‚≠ê 3. RIGHT Function

### ‚úî Syntax  
`=RIGHT(text, number_of_characters)`

### ‚úî Example  
`=RIGHT(C2,3)`

---

# ‚≠ê 4. MID Function

### ‚úî Syntax  
`=MID(text, start_position, number_of_characters)`

### ‚úî Example  
`=MID(A2,2,5)`

---

# ‚≠ê 5. LEN Function

### ‚úî Syntax  
`=LEN(text)`

### ‚úî Example  
`=LEN(B2)`

---

# ‚≠ê 6. TRIM Function

### ‚úî Syntax  
`=TRIM(text)`

### ‚úî Example  
`=TRIM(A2)`

---

# ‚≠ê 7. CONCAT & TEXTJOIN Functions

### ‚úî CONCAT Syntax  
`=CONCAT(text1, text2)`

### ‚úî TEXTJOIN Syntax  
`=TEXTJOIN(delimiter, ignore_empty, range)`

### **Q1. Extract first 4 characters from Full Name.**
Formula:  
`=LEFT(A2,4)`

Answer:

### **Q2. Extract last 3 characters from City.**
Formula:  
`=RIGHT(C2,3)`

Answer:  
uru

### **Q3. Extract 5 characters starting from position 2 in Full Name.**
Formula:  
`=MID(A2,2,5)`

### **Q4. Find total length of Email ID.**
Formula:  
`=LEN(B2)`

### **Q5. Convert City to uppercase.**
Formula:

### **Q6. Convert City to proper case.**

Formula:

Formula:

# ‚≠ê 10. Interview Questions

---

### **Q1. What are text functions in Excel?**  
Functions used to manipulate, clean, and format text data.

---

### **Q2. Difference between LEFT, RIGHT, and MID?**  
LEFT extracts from start, RIGHT from end, MID from middle.

---

### **Q3. Does LEN count spaces?**  
Yes, spaces are counted as characters.

---

### **Q4. Why is TRIM important in data analysis?**  
It removes unwanted spaces that affect matching and accuracy.

---

### **Q5. Difference between CONCAT and TEXTJOIN?**  
TEXTJOIN allows delimiters and ignores empty cells.

---

### **Q6. Can text functions be nested?**  
Yes, multiple text functions can be combined.

---

### **Q7. How do you extract domain from an email?**  
Using RIGHT, LEN, and FIND functions.

---

### **Q8. What does PROPER function do?**  
Capitalizes the first letter of each word.

---

### **Q9. What happens if text is empty in TEXTJOIN?**  
It can be ignored using TRUE parameter.

---

### **Q10. Why are text functions important for Data Analysts?**  
They help clean and standardize textual data for analysis.
