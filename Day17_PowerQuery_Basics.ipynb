# ‚≠ê 1. Introduction to Power Query

Power Query is Excel‚Äôs **data extraction, transformation, and loading (ETL) tool**.
It is used to:
- Import data from multiple sources
- Clean messy datasets
- Transform data without formulas
- Automate repetitive data preparation tasks

Power Query works on the principle of:
**Source ‚Üí Steps ‚Üí Output**

Every action is recorded as a step, making transformations repeatable and auditable.

---

# ‚≠ê 2. Why Power Query is Important

Power Query is preferred over formulas when:
- Data is large
- Data structure changes frequently
- Cleaning steps need to be repeated regularly

Benefits:
- No manual cleaning every time
- Faster than formulas for large data
- Keeps raw data unchanged
- Easy to refresh with new data

This makes Power Query essential for analysts and BI professionals.

---

# ‚≠ê 3. Accessing Power Query

You can open Power Query from:
- Data ‚Üí Get Data
- Data ‚Üí From Table/Range

When data is loaded into Power Query:
- Excel opens the **Power Query Editor**
- Original data remains unchanged
- Transformations happen in a separate environment

---

# ‚≠ê 4. Power Query Editor Interface

### Main Components

- **Ribbon**  
  Contains transformation commands like remove columns, split, merge, etc.

- **Queries Pane (Left)**  
  Shows all queries in the workbook.

- **Data Preview (Center)**  
  Displays a sample of the data.

- **Applied Steps (Right)**  
  Shows all transformations applied in sequence.

Each step can be:
- Edited
- Reordered
- Deleted

This makes Power Query very flexible.

---

# ‚≠ê 5. Loading Data Using Power Query

Common data sources:
- Excel files
- CSV / Text files
- Databases
- Web data
- Folder (multiple files)

Loading data from a folder is especially useful when:
- Files are added regularly
- Structure is consistent

Power Query can combine files automatically.

---

# ‚≠ê 6. Basic Data Cleaning Operations

Power Query provides point-and-click data cleaning:

### Remove Columns
- Remove unnecessary columns to reduce file size

### Rename Columns
- Standardize column names

### Remove Blank Rows
- Clean empty or irrelevant rows

### Remove Duplicates
- Ensure unique records

### Change Data Types
- Convert text to number/date
- Critical for calculations

Incorrect data types are a common source of errors.

---

# ‚≠ê 7. Handling Errors & Missing Data

Power Query allows:
- Removing rows with errors
- Replacing error values
- Replacing null (missing) values

This ensures clean datasets before analysis.

---

# ‚≠ê 8. Applied Steps Concept (Very Important)

Each transformation is stored as a **step**.
Power Query executes steps sequentially.

Advantages:
- Easy debugging
- Reusable logic
- Automated refresh

This is why Power Query is considered a low-code tool.
