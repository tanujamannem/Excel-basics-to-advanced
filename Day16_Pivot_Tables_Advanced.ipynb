# ‚≠ê 1. Introduction to Advanced Pivot Tables

Advanced Pivot Table features are used when basic totals are not enough.
They help analysts:
- Create custom business metrics
- Summarize data at higher levels
- Add interactivity for users
- Build dashboards and management reports

These features reduce dependency on complex formulas and make reports easier to maintain.

---

# ‚≠ê 2. Calculated Fields

Calculated Fields allow you to create **new measures inside a Pivot Table** using existing numeric fields.

Key points:
- Calculated fields work on **aggregated data**
- They are evaluated after the Pivot Table groups the data
- They exist only inside the Pivot Table (not in source data)

Common use cases:
- Profit = Sales ‚àí Cost
- Margin = Profit / Sales
- Revenue per Unit = Sales / Quantity

### How Calculated Fields Work
If a Pivot Table groups Sales by Region:
- Sales is already summed
- Cost is already summed
- Calculated Field operates on these totals

This makes calculated fields ideal for summary-level analysis.

---

# ‚≠ê 3. Grouping Data

Grouping allows you to analyze data at a **higher or more meaningful level**.

### Date Grouping
You can group dates into:
- Months
- Quarters
- Years

This helps answer questions like:
- Monthly sales trends
- Quarterly performance
- Year-over-year comparison

### Number Grouping
Numeric values can be grouped into ranges:
- Sales ranges
- Quantity buckets
- Salary slabs

Example:
- 0‚Äì3000
- 3001‚Äì7000
- 7000+

This simplifies analysis of large numeric datasets.

### Manual Grouping
You can manually group items:
- Combine multiple products
- Merge similar categories

Grouping helps reduce complexity and improve readability.

---

# ‚≠ê 4. Slicers

Slicers are **visual filtering tools** that make Pivot Tables interactive.

Advantages of slicers:
- Easy to use (no dropdowns)
- Clearly show applied filters
- Suitable for dashboards

Slicers can filter:
- Regions
- Categories
- Time periods

### Multiple Pivot Control
One slicer can control:
- Multiple Pivot Tables
- Pivot Tables from the same data source

This ensures consistency across reports.

---

# ‚≠ê 5. Pivot Charts

Pivot Charts visually represent Pivot Table results.

Key characteristics:
- Always linked to Pivot Tables
- Update automatically when data changes
- Respect slicers and filters

Pivot Charts are commonly used in:
- Dashboards
- Management reports
- Executive summaries

Unlike normal charts, Pivot Charts depend entirely on Pivot Table structure.
