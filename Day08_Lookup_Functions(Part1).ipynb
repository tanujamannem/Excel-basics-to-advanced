# ‚≠ê 1. Introduction to Lookup Functions

Lookup functions are used to **search for a value in a table and return a related value**.

Real-life guarantee:
- Used in almost EVERY Excel job
- Asked in ALL data analyst interviews
- Foundation for advanced lookups (XLOOKUP, INDEX+MATCH)

Today you will learn:
- VLOOKUP
- HLOOKUP
- Exact vs Approximate match
- Common errors & limitations

---

# ‚≠ê 2. What is VLOOKUP?

VLOOKUP = **Vertical Lookup**

It searches:
- First column of a table
- Finds a match
- Returns value from another column in the SAME ROW

---

# ‚≠ê 3. VLOOKUP Syntax (Very Important)

`=VLOOKUP(lookup_value, table_array, col_index_num, [range_lookup])`

Explanation:
- lookup_value ‚Üí value you are searching for
- table_array ‚Üí full data range
- col_index_num ‚Üí column number to return
- range_lookup ‚Üí FALSE (Exact), TRUE (Approximate)

---

# ‚≠ê 4. VLOOKUP Example Explained

Formula:
`=VLOOKUP("P103", A2:D11, 2, FALSE)`

Meaning:
- Search P103
- In column A
- Return value from column 2 (Product)
- Exact match only

---

# ‚≠ê 5. What is HLOOKUP?

HLOOKUP = **Horizontal Lookup**

Used when:
- Data is arranged horizontally
- Headers are in the FIRST ROW

Syntax:
`=HLOOKUP(lookup_value, table_array, row_index_num, [range_lookup])`

Formula:

Formula:

Answer:

### **Q3. Find Price for Product ID P110.**
Formula:

Formula:

### **Q5. What happens if column index is wrong?**
Formula:  
`=VLOOKUP("P101",A2:D11,6,FALSE)`

Answer:

---

### **Q6. Use VLOOKUP with approximate match (range_lookup = TRUE).**
Formula:  
`=VLOOKUP(100,D2:D11,1,TRUE)`

Answer:  
Returns nearest lower match

---

### **Q7. Why FALSE is preferred in interviews?**
Answer:  
Ensures exact match and avoids wrong results.

---

### **Q8. Can VLOOKUP search to the LEFT?**
Answer:  
No (major limitation)

---

### **Q9. What error if value not found?**
Answer:

---

### **Q10. How to fix #N/A error?**
Answer:  
Use IFERROR with VLOOKUP.

# ‚≠ê 8. Common VLOOKUP Limitations

- Cannot lookup left
- Breaks if column order changes
- Slower on large datasets
- One-column lookup only

---

# ‚≠ê 9. Interview Questions (10)

---

### **Q1. What does VLOOKUP do?**  
Searches vertically and returns matching value.

---

### **Q2. What is col_index_num?**

No.
