# ‚≠ê 1. Introduction to Excel Tables

Excel Tables convert normal data ranges into **dynamic, structured objects**.

Using Tables helps in:
- Automatic expansion of formulas
- Cleaner formulas (structured references)
- Dynamic charts & pivots
- Professional data modeling

Shortcut to create a table:
Ctrl + T

---

# ‚≠ê 2. Creating an Excel Table

### ‚úî Steps
1. Select any cell inside data
2. Press Ctrl + T
3. Check "My table has headers"
4. Click OK

Excel automatically:
- Adds filters
- Applies formatting
- Names the table

---

# ‚≠ê 3. Table Features

### ‚úî Key Features
- Automatic formula fill
- Dynamic range expansion
- Structured references
- Built-in filters & sorting
- Table styles

---

# ‚≠ê 4. Structured References

Structured references replace cell ranges with column names.

### ‚úî Example
Normal formula:
`=SUM(C2:C11)`

Table formula:
`=SUM(Table1[Salary])`

Advantages:
- Easy to read
- No range break when rows are added

---

# ‚≠ê 5. Calculated Columns

When a formula is entered in one row of a table:
- Excel automatically fills the entire column

### ‚úî Example
`=[@Salary]*0.10`

Applied to all rows automatically.

# ‚≠ê 9. Interview Questions

---

### **Q1. What is an Excel Table?**  
A structured, dynamic range with built-in features.

---

### **Q2. Shortcut to create a table?**  
Ctrl + T

---

### **Q3. What are structured references?**  
Formulas that use column names instead of cell ranges.

They expand automatically and prevent formula breaks.

Yes, highly recommended.

Yes, formulas auto-fill entire column.

Table Design ‚Üí Table Name box.

---

### **Q9. Can structured references be used outside tables?**  
Yes, to refer to table data.

### **Q10. When should you avoid tables?**  
Rarely ‚Äî mainly when exporting to systems that don‚Äôt support tables.
