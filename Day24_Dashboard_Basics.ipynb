# ‚≠ê 3. Preparing Data for Dashboard

### ‚úî Step 1: Convert Data to Table
1. Select entire dataset
2. Press **Ctrl + T**
3. Check **My table has headers**
4. Click OK

Why:
- Tables auto-expand
- Works better with Pivot Tables

# ‚≠ê 7. Adding Slicers (Interactivity)

Slicers allow filtering dashboard visually.

---

## üîπ Add Region Slicer

### ‚úî Steps
1. Click any pivot table
2. Go to **PivotTable Analyze ‚Üí Insert Slicer**
3. Select **Region**
4. Click OK

---

## üîπ Add Category Slicer

Repeat steps ‚Üí select **Category**

---

## üîπ Link Slicers to ALL Pivots

### ‚úî VERY IMPORTANT
1. Click slicer
2. Go to **Slicer ‚Üí Report Connections**
3. Check ALL pivot tables
4. Click OK

Now slicer controls entire dashboard.

---

# ‚≠ê 8. Dashboard Layout Best Practices

---

## üîπ Layout Rules

- KPIs at top
- Charts in middle
- Slicers on left or right
- White background
- Consistent font

---

## üîπ Alignment

### ‚úî Steps
1. Select multiple objects
2. Go to **Shape Format ‚Üí Align**
3. Align Top / Distribute Horizontally

---

## üîπ Remove Distractions

- Hide gridlines
- Remove pivot field buttons
- Lock pivot sheets if needed
