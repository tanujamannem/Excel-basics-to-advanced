# ‚≠ê 1. Introduction to Conditional Functions

Conditional functions perform calculations or return results **based on specific criteria**.

They are widely used in data analysis to:
- Segment data
- Apply business rules
- Perform conditional aggregations
- Handle errors gracefully
- Highlight important patterns visually

---

# ‚≠ê 2. IFERROR Function

IFERROR is used to handle errors in formulas and replace them with a meaningful value.

### ‚úî Syntax  
`=IFERROR(value, value_if_error)`

### ‚úî Example  
`=IFERROR(A2/B2, "Invalid")`

If the formula results in an error, Excel returns **"Invalid"** instead of an error code.

---

# ‚≠ê 3. COUNTIF Function

COUNTIF counts the number of cells that meet a given condition.

### ‚úî Syntax  
`=COUNTIF(range, criteria)`

### ‚úî Example  
`=COUNTIF(B2:B11, "Completed")`

---

# ‚≠ê 4. SUMIF Function

SUMIF adds values that satisfy a condition.

### ‚úî Syntax  
`=SUMIF(range, criteria, sum_range)`

### ‚úî Example  
`=SUMIF(A2:A11, "Stationery", C2:C11)`

---

# ‚≠ê 5. AVERAGEIF Function

AVERAGEIF calculates the average of values that meet a condition.

### ‚úî Syntax  
`=AVERAGEIF(range, criteria, average_range)`

### ‚úî Example  
`=AVERAGEIF(B2:B11, "Kitchen", D2:D11)`

---

# ‚≠ê 6. Conditional Formatting Rules

Conditional Formatting allows visual highlighting of data based on rules.

### ‚úî Examples of Rules  
- Highlight cells greater than a value (e.g., >100 ‚Üí red fill)  
- Highlight duplicate values  
- Highlight text containing a keyword  
- Use formulas for custom conditions

---

# ‚≠ê 7. Color Scales

Color scales apply a gradient of colors to a range of numbers to visualize differences.

### ‚úî Steps  
1. Select range (e.g., D2:D11)  
2. Home ‚Üí Conditional Formatting ‚Üí Color Scales  
3. Choose a two-color or three-color scale  
4. Higher values ‚Üí darker shade, lower values ‚Üí lighter shade

### ‚úî Example  
Price range 5‚Äì700 ‚Üí Light green (low) to Dark green (high)

---

# ‚≠ê 8. Icon Sets

Icon sets use visual icons to represent values in a range.

### ‚úî Steps  
1. Select range (e.g., D2:D11)  
2. Home ‚Üí Conditional Formatting ‚Üí Icon Sets  
3. Choose an icon style (arrows, traffic lights, stars, etc.)  
4. Excel automatically applies icons based on relative value

### ‚úî Example  
- Top 3 prices ‚Üí green up arrow  
- Middle prices ‚Üí yellow sideways arrow  
- Lowest prices ‚Üí red down arrow

# ‚≠ê 11. Conditional Formatting (Concept)

Conditional Formatting visually highlights cells based on rules.

Common use cases:
- Highlight high/low values
- Flag errors or exceptions
- Show trends using colors
- Improve dashboard readability

Examples:
- Quantity > 50 ‚Üí Green
- Price > 100 ‚Üí Red
- Top 10 values
- Data bars and color scales

# ‚≠ê 12. Interview Questions

---

### **Q1. What does IFERROR do?**  
Replaces error values with a specified output.

---

### **Q2. Difference between COUNT and COUNTIF?**  
COUNT counts numbers; COUNTIF counts based on criteria.

---

### **Q3. When do you use SUMIF?**  
When summing values conditionally.

---

### **Q4. Can SUMIF handle multiple conditions?**  
No, use SUMIFS for multiple conditions.

---

### **Q5. What happens if no criteria match in SUMIF?**  
Result is 0.

---

### **Q6. Difference between AVERAGEIF and AVERAGE?**  
AVERAGEIF applies a condition.

---

### **Q7. What is Conditional Formatting used for?**  
To visually highlight data based on rules.

---

### **Q8. Does Conditional Formatting change data?**  
No, only visual appearance.

---

### **Q9. Can IFERROR hide all Excel errors?**  
Yes, if applied correctly.

---

### **Q10. Why are conditional functions important in analysis?**  
They help segment, validate, and summarize data.
