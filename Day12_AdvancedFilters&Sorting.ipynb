# ‚≠ê 2. Sorting Data (Basic to Advanced)

Sorting arranges data in a specific order.

### ‚úî Types of Sorting
- Ascending (A ‚Üí Z, Small ‚Üí Large)
- Descending (Z ‚Üí A, Large ‚Üí Small)

### ‚úî Example
Sort sales data by Price (High to Low).

Steps:
1. Select any cell in data
2. Go to **Data ‚Üí Sort**
3. Choose column ‚Üí Order ‚Üí Descending

---

# ‚≠ê 3. Multi-Level Sorting

Used when sorting by more than one column.

### ‚úî Example
Sort data by:
1. Category (A ‚Üí Z)
2. Price (High ‚Üí Low)

Steps:
1. Data ‚Üí Sort
2. Add Level
3. First: Category (A ‚Üí Z)
4. Second: Price (Largest ‚Üí Smallest)

This is VERY common in business datasets.

# ‚≠ê 4. Custom Sorting

### ‚úî Example

# ‚≠ê 5. Filtering Data (AutoFilter)

Filtering hides rows that do not meet criteria.

### ‚úî Enable Filter
Shortcut: **Ctrl + Shift + L**

### ‚úî Common Filters
- Text filters (Contains, Begins with)
- Number filters (> , < , Between)
- Date filters (This Month, Last Year)
- Filter by color

---

# ‚≠ê 6. Advanced Filtering (Powerful Feature)

Advanced Filter is used when:
- Multiple complex conditions
- AND / OR logic
- Extracting filtered results to a new location
- Removing duplicates with conditions

# ‚≠ê 9. Difference: Filter vs Advanced Filter

Filter:
- Simple conditions
- Hides rows
- Works on same dataset

Advanced Filter:
- Complex logic (AND / OR)
- Extracts data
- Can remove duplicates
