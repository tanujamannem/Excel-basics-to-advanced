# ‚≠ê 1. Introduction to Macros

Macros are used to **automate repetitive tasks** in Excel.

A macro:
- Records your actions
- Converts them into VBA code
- Replays them anytime with one click

Macros are useful for:
- Formatting reports
- Cleaning data repeatedly
- Saving time
- Improving productivity

Manual tasks:

# ‚≠ê 5. Record a Macro (Step-by-Step)

### Step 1: Start Recording
1. Go to **Developer ‚Üí Record Macro**
2. Macro name: `Format_Report`
3. Shortcut key: Ctrl + Shift + F (optional)
4. Store macro in: **This Workbook**
5. Click OK

---

### Step 2: Perform Actions (Recording ON)

1. Select header row
   - Make **Bold**
2. Select entire table
   - Apply **All Borders**
3. Select Salary & Bonus columns
   - Home ‚Üí Number ‚Üí Currency
4. Apply filter
   - Data ‚Üí Filter
5. Highlight IT department
   - Select Department column
   - Conditional Formatting ‚Üí Highlight Cells ‚Üí Equal To ‚Üí IT

---

### Step 3: Stop Recording
1. Developer ‚Üí **Stop Recording**

Macro is now saved.

---

# ‚≠ê 6. Run the Macro

### Method 1:
Developer ‚Üí Macros ‚Üí Format_Report ‚Üí Run

### Method 2:
Use shortcut key (Ctrl + Shift + F)

Excel repeats all steps automatically.

# ‚≠ê 8. Absolute vs Relative Macro Recording

---

### Absolute Recording (Default)
- Works only for same cells (A1:E6)

---

### Relative Recording
Allows macro to work on **any selected range**

### How to Enable:
Developer ‚Üí Use Relative References ‚Üí Record Macro

# ‚≠ê 9. Practice Questions

Q1. What is a macro?  
Answer: Recorded set of actions.

Q2. Where are macros stored?  
Answer: In workbook or personal macro workbook.

Q3. Can macros save time?  
Answer: Yes.

Answer: No (recording is enough).

Answer: VBA.

Q7. Can macros be edited?  
Answer: Yes (in VBA editor).

Answer: Yes.

Q9. Are macros allowed in all companies?  
Answer: Depends on security policy.

Q10. File extension for macro workbook?
