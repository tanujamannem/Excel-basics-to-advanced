# ‚≠ê 1. Introduction to Number Functions

Number functions are used to **control decimals, round values, extract integers, and handle remainders**.

They are commonly used in:
- Financial calculations
- Billing systems
- Data standardization
- Interview problem solving

Today you will learn:
- ROUND
- ROUNDUP
- ROUNDDOWN
- INT
- MOD

# ‚≠ê 9. Interview Questions

---

### **Q1. Difference between ROUND and ROUNDUP?**  
ROUND follows standard rounding rules, ROUNDUP always rounds upward.

---

### **Q2. What does ROUNDDOWN do?**  
Always rounds toward zero.

---

### **Q3. Difference between INT and ROUNDDOWN?**  
INT rounds down to nearest integer, even for negatives.

---

### **Q4. What does MOD return?**  
The remainder after division.

---

### **Q5. Where is MOD used in real scenarios?**  
Even/odd checks, cyclic patterns, grouping.

---

### **Q6. Can ROUND handle negative num_digits?**  
Yes, to round to tens, hundreds, etc.

---

### **Q7. What happens if num_digits = 0?**  
Rounds to nearest integer.

---

### **Q8. Why are number functions important in analysis?**  
They standardize numeric data for reporting.

---

### **Q9. Can MOD return zero?**  
Yes, when number is perfectly divisible.

---

### **Q10. Are number functions used in financial models?**  
Yes, heavily used for currency and totals.
