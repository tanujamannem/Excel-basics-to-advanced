## üîπ Method 1: Remove Errors Using IFERROR

Used when:
- Calculations fail
- You want safe output

### ‚úî Formula

Meaning:
- If error ‚Üí show 0
- Else ‚Üí show result

---

## üîπ Method 2: Filter & Delete Error Rows

### ‚úî Steps
1. Select your dataset
2. Go to **Data ‚Üí Filter**
3. Click dropdown on column
4. Select **Error values**
5. Delete selected rows

Used when:
- Error rows are invalid

---

## üîπ Method 3: Power Query Error Removal

### ‚úî Steps
1. Select data
2. Go to **Data ‚Üí From Table/Range**
3. Power Query Editor opens
4. Right-click column ‚Üí **Remove Errors**
5. Click **Close & Load**

Best for:
- Large datasets
- Repeated cleaning

# ‚≠ê 4. Regular Expressions (Power Query) ‚Äî Advanced Cleaning

Regex = Pattern-based text cleaning  
(Used in professional projects)

---

## üîπ Where Regex Works in Excel

Only in:
Power Query

---

## üîπ How to Open Power Query

1. Select data
2. Go to **Data ‚Üí From Table/Range**
3. Power Query Editor opens

---

## üîπ Example 1: Remove Numbers from Text

Column value:
Asha123

### ‚úî Steps
1. Select column
2. Go to **Transform ‚Üí Replace Values**
3. Click **Advanced options**
4. Enable **Use Regex**
5. Find: `[0-9]`
6. Replace with: (blank)
7. OK

Result:
Asha

---

## üîπ Example 2: Extract Only Numbers

Text:
INV-5678

### ‚úî Steps
1. Select column
2. Go to **Add Column ‚Üí Extract ‚Üí Text After Delimiter**
(Or use Regex)
3. Regex pattern: `\d+`

Result:
5678

---

## üîπ Example 3: Remove Special Characters

Pattern:
[^A-Za-z ]

Meaning:
- Keep only letters & space
