# ‚≠ê 1. Introduction to Date & Time Functions

Date & Time functions are used to **calculate durations, extract date parts, and track timelines**.

They are extremely important for:
- Experience calculation
- Age calculation
- SLA tracking
- Reporting & dashboards

Today you will learn:
- TODAY
- NOW
- DAY, MONTH, YEAR
- DATEDIF
- EOMONTH

---

# ‚≠ê 2. TODAY Function

### ‚úî Syntax

### ‚úî Example  
Returns current system date  
`=TODAY()`

---

# ‚≠ê 3. NOW Function

### ‚úî Syntax  
`=NOW()`

### ‚úî Example  
Returns current date and time  
`=NOW()`

---

# ‚≠ê 4. DAY, MONTH, YEAR Functions

### ‚úî Syntax  
`=DAY(date)`  
`=MONTH(date)`  
`=YEAR(date)`

### ‚úî Example  
`=DAY(A2)`  
`=MONTH(A2)`  
`=YEAR(A2)`

---

# ‚≠ê 5. DATEDIF Function

### ‚úî Syntax  
`=DATEDIF(start_date, end_date, "unit")`

Units:  
"Y" ‚Üí Years  
"M" ‚Üí Months  
"D" ‚Üí Days  

### ‚úî Example  
`=DATEDIF(A2,B2,"Y")`

---

# ‚≠ê 6. EOMONTH Function

### ‚úî Syntax  
`=EOMONTH(start_date, months)`

### ‚úî Example  
`=EOMONTH(A2,0)`

# ‚≠ê 9. Interview Questions

---

### **Q1. What does TODAY() return?**  
Current system date.

---

### **Q2. Difference between TODAY and NOW?**  
TODAY ‚Üí date only  
NOW ‚Üí date and time

---

### **Q3. What does DATEDIF function do?**  
Calculates difference between two dates.

---

### **Q4. Is DATEDIF visible in Excel function list?**  
No, but it works when typed manually.

---

### **Q5. What are common DATEDIF units?**  
"Y", "M", "D".

---

### **Q6. What does EOMONTH return?**  
Last day of a given month.

---

### **Q7. Can Excel store dates as numbers?**  
Yes, dates are stored as serial numbers.

---

### **Q8. Why are Date functions important for analysts?**  
They help track time-based trends and durations.

---

### **Q9. How do you calculate age in Excel?**  
Using DATEDIF with "Y".

---

### **Q10. Can Date & Time functions be used in dashboards?**  
Yes, they are widely used in KPIs and reports.
