# ‚≠ê 1. Introduction to Logical Functions

Logical functions are used to **test conditions and return TRUE/FALSE or custom results**.

They are widely used in:
- Business rules
- Decision making
- Categorization
- Interview problem solving

Today you will learn:
- IF
- AND
- OR
- NOT
- Nested IF

---

# ‚≠ê 2. IF Function

### ‚úî Syntax  
`=IF(condition, value_if_true, value_if_false)`

### ‚úî Examples  
`=IF(A2>=18,"Adult","Minor")`  
`=IF(B2>80,"Pass","Fail")`

---

# ‚≠ê 3. AND Function

### ‚úî Syntax  
`=AND(condition1, condition2, ...)`

### ‚úî Example  
`=AND(A2>=18, B2>=60)`

---

# ‚≠ê 4. OR Function

### ‚úî Syntax  
`=OR(condition1, condition2, ...)`

### ‚úî Example  
`=OR(B2>90, C2="Yes")`

---

# ‚≠ê 5. NOT Function

### ‚úî Syntax  
`=NOT(condition)`

### ‚úî Example  
`=NOT(A2>50)`

---

# ‚≠ê 6. Nested IF

### ‚úî Syntax  
`=IF(condition1, result1, IF(condition2, result2, result3))`

### ‚úî Example  
`=IF(B2>80,"A",IF(B2>60,"B","C"))`

# ‚≠ê 9. Interview Questions

---

### **Q1. What are logical functions in Excel?**  
Functions that test conditions and return TRUE/FALSE or custom results.

---

### **Q2. Difference between AND and OR?**  
AND ‚Üí all conditions must be TRUE  
OR ‚Üí any one condition must be TRUE

---

### **Q3. What does NOT function do?**  
Reverses TRUE to FALSE and FALSE to TRUE.

---

### **Q4. Can IF return text values?**  
Yes, numbers, text, or formulas.

---

### **Q5. What is Nested IF?**  
Multiple IF functions combined to test many conditions.

---

### **Q6. What is the limitation of Nested IF?**  
Becomes complex and hard to read for many conditions.

---

### **Q7. Can logical functions be combined?**  
Yes, IF can be combined with AND, OR, NOT.

---

### **Q8. What does AND return if one condition fails?**  
FALSE.

---

### **Q9. What happens if IF condition is FALSE?**  
It returns the value_if_false argument.

---

### **Q10. Why are logical functions important for Data Analysts?**  
They help apply business rules and classify data accurately.
